```python
def transform_csv_to_midi(file_path: str,name_output:str):
    to_convert = np.loadtxt(file_path)
    output = pyd.PrettyMIDI(resolution=127)
    resolution = 100
    pitchs = np.arange(23,93)
    n = len(to_convert[0])
    instru = pyd.Instrument(0)
    for i in range(len(pitchs)):
        to_analyze = to_convert[i]
        to_analyze_trick = np.concatenate([to_analyze,[0]])-np.concatenate([[0],to_analyze])
        starting_time = np.argwhere(to_analyze_trick==1).astype(float)/1000#resolution
        ending_time = np.argwhere(to_analyze_trick==-1).astype(float)/1000#resolution
        for start,end in zip(starting_time,ending_time):
            instru.notes.append(pyd.Note(100,pitchs[i],start[0],end[0]))
    output.instruments.append(instru)
    output.write("Generated_midi/"+name_output+".mid")
```
